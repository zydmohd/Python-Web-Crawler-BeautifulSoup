# This is examole for building web crawler within given Wikipedia link , then the app will search for all links inside the page and choose one randomly to open and repeat the cycle.
# It is practic example of use BeautifulSoup .

```python
# Import the required libraries 
# urlopen : open url
# re : to use regex
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import random
```

```python
# Creat function :
# Open URL ( urlLink)
# Creat BeautifulSoup object
# Creat search template by regex,(choose only the links stars with wiki and exclude the rest)
# Filter the links , creat Dictionary has the result values
# return the list
def getlinks(urlLink):
    
    html1 = urlopen(urlLink)
    bs1 = BeautifulSoup(html1,'html.parser')
    hrefText1 = re.compile(r'^(/wiki/)[^(Category)][^(Help:)](.+)')
    linkList1 = bs1.find_all('a', href = hrefText1 , class_='', accesskey='')
    return linkList1
```

```python
# I choosed 1 link to start with
url1 = 'http://en.wikipedia.org/wiki/Kevin_Bacon'
```

```python
# Get all links in list
linkList1 = getlinks(url1)
```

```python
# Choose 1 link randomly from the list
def randomXValue():
    x = random.randint(0,len(linkList1))
    return x
```

```python
#Test result

print(randomXValue())
```

```python
# Choose the corrosponding link from the list
choosenLink=linkList1[randomXValue()].get('href')
print(choosenLink)
```

```python
# Cycle 10 time , page>> choose link>> open link>> choose random link>> repeat..
i = 10
while i > 0 :
    i=i-1
    newLink='http://en.wikipedia.org'+ linkList1[randomXValue()].get('href')
    print(newLink)
    linkList1= getlinks(newLink)
    
    

    
```
